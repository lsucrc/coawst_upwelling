Import libraries and define settings,use the Dataset class of the netCDF4 module.

```python
from netCDF4 import Dataset
import numpy as np
```

Start by opening the file that contains the variables we want to eventually plot. fh becomes the file handle of the open netCDF file, and the 'r' denotes that we want to open the file in read only mode.

```python
nc_file = 'ocean_his.nc'
fh = Dataset(nc_file, mode='r')
```

Read the data from any of the variables contained in fh.

```python
x_rho = fh.variables['x_rho'][:]
y_rho= fh.variables['y_rho'][:]
temp = fh.variables['temp'][0][0]
```

Close the file.

```python
fh.close()
```

Plotting data using matplotlib.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.figure(figsize=(7,4))
plt.title("Temperature ($^\circ$C)")
plt.contourf(x_rho,y_rho,temp)
plt.colorbar()
plt.show()
plt.savefig("Temperature.png")
```
